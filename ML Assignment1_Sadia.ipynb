```python
def line(m,c,x):
    x= np.arange(0, x)
    y = m*x + c
    return x,y
```

```python
def line1(w):
    m= -w[1]/w[2]
    c= -w[0]/w[2]
    x,y= line(m,c,30)
    return x,y
```

```python
random.seed(90)

x1 = [random.random()*30 for i in range(0,20)]
y1 = [random.random()*30 for i in range(0,20)]

m = 1
c = .5
x,y=line(m,c,30)

X1,Y1,X2,Y2=([] for i in range(4))

for i in range(20):
    h = y1[i] - (m*x1[i] + c)
    if  h>= 0:
        X1.append(x1[i])
        Y1.append(y1[i])
    else:
        X2.append(x1[i])
        Y2.append(y1[i])  

plt.plot(x, y, color='red')
plt.scatter(X1, Y1, c = "blue")
plt.scatter(X2, Y2, c = "green")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
def perceptron(n_list, d, limit):
    w_list=[]
    for i in range(d+1):
        w_list.append(0.0)
    count=0
    update=0
    
    while count <100:
        for i in range(limit):
            x_i=n_list[i]
            activation=0
            for j in range(d+1):
                activation= activation+w_list[j]*x_i[j]
                
                if activation >= 0:
                    y_pred = 1
                else:
                    y_pred = -1
                
                if y_pred != x_i[d+1]:
                    w_list=[(w_list[k] + x_i[d+1]* x_i[k]) for k in range(d+1)]
                    update= update+1
        count=count+1
    print("Update:", update)
    return(w_list)          
  
```

```python
random.seed(90)

x1 = [random.random()*30 for i in range(0,20)]
y1 = [random.random()*30 for i in range(0,20)]

m = 1
c = .5
x,y=line(m,c,30)

X1,Y1,X2,Y2,w,n=([] for i in range(6))

for i in range(20):
    h = y1[i] - (m*x1[i] + c)
    if  h>= 0:
        X1.append(x1[i])
        Y1.append(y1[i])
        n.append([1,x1[i],y1[i],1])
    else:
        X2.append(x1[i])
        Y2.append(y1[i])
        n.append([1,x1[i],y1[i],-1])  

limit=20
w= perceptron(n,2,limit)
x3,y3= line1(w)

plt.plot(x3, y3, '--',label='hypothesis(g)', color='yellow')
plt.plot(x, y,'-r',label='target function(f)', color='red')
plt.scatter(X1, Y1, c = "blue")
plt.scatter(X2, Y2, c = "green")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```
