```python
def line(m,c):
    x= np.arange(0, 30)
    y = m*x + c
    return x,y
```

```python
import random
from random import seed
from matplotlib import pyplot as plt
import numpy as np

random.seed(50)

x1 = [random.random()*30 for i in range(0,20)]
y1 = [random.random()*30 for i in range(0,20)]

m = 1
c = .5
x,y=line(m,c)


X1=[]
Y1=[]
X2=[]
Y2=[]

for i in range(20):
    h = y1[i] - (m*x1[i] + c)
    if  h>= 0:
        X1.append(x1[i])
        Y1.append(y1[i])
    else:
        X2.append(x1[i])
        Y2.append(y1[i])  

plt.plot(x, y, color='red')
plt.scatter(X1, Y1, c = "blue")
plt.scatter(X2, Y2, c = "green")
plt.legend()
plt.show()
```

```python
def perceptron(n_list, d, limit):
    w_list=[]
    for i in range(d+1):
        w_list.append(0.0)
    count=0
    
    while count <limit:
        for i in range(20):
            x_i=n_list[i]
            activation=0
            for j in range(d+1):
                activation= activation+w_list[j]*x_i[j]
                
                if activation >= 0:
                    y_pred = 1
                else:
                    y_pred = -1
                
                if y_pred != x_i[d+1]:
                    w_list=[(w_list[k] + x_i[d+1]* x_i[k]) for k in range(d+1)]
        count=count+1

    return(w_list)          
  
```
