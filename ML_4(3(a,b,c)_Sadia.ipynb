```python
def data():
    rows = []
    with open("/content/drive/MyDrive/ML/features.train.txt",'r') as data_file:
        for line in data_file:
            data = line.split()
            rows.append(data)
        rows = np.array(rows)
        df = pd.DataFrame(rows)
        df.columns =['col1', 'col2', 'col3']
        #print(df)
    
    df['col1'] = df['col1'].astype(float)
    df['col2'] = df['col2'].astype(float)
    df['col3'] = df['col3'].astype(float)
    df['col1'] = df['col1'].astype(int)
    df['col1'] = np.where(df['col1'] == 1, 1,-1)
    df = df.sample(frac = 1)
    return df
    
```

```python
df=data()
print(df)
```

```python
def train_test(df):
    train=df.iloc[:500]
    test=df.iloc[500:]
    X_train=train[["col2","col3"]].to_numpy()
    y_train=train[["col1"]].to_numpy()
    X_test=test[["col2","col3"]].to_numpy()
    y_test=test[["col1"]].to_numpy()
    return X_train,X_test,y_train,y_test
    
```

```python
X_train,X_test,y_train,y_test= train_test(df)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
def KNN(X_train,X_test,y_train,y_test):
    knn = KNeighborsRegressor(n_neighbors=3)
    knn.fit(X_train, y_train)
    
    y_pred = knn.predict(X_train)
    msein = mean_squared_error(y_train, y_pred)
    rmsein = sqrt(msein)
    k_ein = float("{:.2f}".format(rmsein))
    
    y_t = knn.predict(X_test)
    mseout = mean_squared_error(y_test, y_t)
    rmseout = sqrt(mseout)
    k_eout = float("{:.2f}".format(rmseout))
    
    return k_ein,k_eout
```

```python
knn_Ein, knn_Eout=([] for i in range(2))
```

```python
for i in range(20):
    df=data()
    X_train,X_test,y_train,y_test= train_test(df)
    knn_ein,knn_eout= KNN(X_train,X_test,y_train,y_test)
    knn_Ein.append(knn_ein)
    knn_Eout.append(knn_eout)
    
```

```python
print(knn_Ein)
print(knn_Eout)
```

```python
def plotErrors(n,e_in,e_out):
    plt.figure(figsize=(4,4))
    plt.plot(range(1,n),e_in,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10,label='E_in')
    plt.plot(range(1,n),e_out,color='blue', linestyle='dashed', marker='x',
         markerfacecolor='orange', markersize=10,label='E_out')
    
    plt.title('Error Rate')
    plt.xlabel('no of iteration')
    plt.ylabel('Error')
    plt.legend(loc="upper left")
    plt.show()
```

```python
plotErrors(21,knn_Ein,knn_Eout)
```

```python
from imblearn.under_sampling import TomekLinks
```

```python
def KNN_CNN(X_train_s,X_test,y_train_s,y_test,X_train,y_train):
    knn = KNeighborsRegressor(n_neighbors=3)
    knn.fit(X_train_s, y_train_s)
    
    y_pred = knn.predict(X_train)
    msein = mean_squared_error(y_train, y_pred)
    rmsein = sqrt(msein)
    k_ein = float("{:.2f}".format(rmsein))
    
    y_t = knn.predict(X_test)
    mseout = mean_squared_error(y_test, y_t)
    rmseout = sqrt(mseout)
    k_eout = float("{:.2f}".format(rmseout))
    
    return k_ein,k_eout
```

```python
def CNN(X_train,X_test,y_train,y_test):
    undersample = TomekLinks()
    X_train_s, y_train_s = undersample.fit_resample(X_train, y_train,)
    cnn_ein,cnn_eout=KNN_CNN(X_train_s,X_test,y_train_s,y_test,X_train,y_train)
    return cnn_ein,cnn_eout
```

```python
cnn_Ein, cnn_Eout=([] for i in range(2))
```

```python
for i in range(20):
    df=data()
    X_train,X_test,y_train,y_test= train_test(df)
    cnn_ein,cnn_eout= CNN(X_train,X_test,y_train,y_test)
    cnn_Ein.append(cnn_ein)
    cnn_Eout.append(cnn_eout)
```
