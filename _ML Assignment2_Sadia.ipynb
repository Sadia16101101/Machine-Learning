```python
import sys
import random
d= int(input())

L=[float(i) for i in input().split()]

G=[float(i) for i in input().split()]  
w=[]
for i in range(d):
    val=(random.randint(L[i], G[i]))
    w.append(val)  

learning_rate= 0.2
count=0
while count <2:
    for i in w:
        print(i, end = " ")
    print()
    sys.stdout.flush()
    fx=input()
    f_x=[float(i) for i in input().split()]
    for i in range (0,d):
        w[i]= w[i]- learning_rate * f_x[i]
    count=count+1    
```

```python
import sys
import random
d= int(input())
L=[]
for i in range(d):
    ele = float(input())
    L.append(ele)


G=[]
for i in range(d):
    ele = float(input())
    G.append(ele)    
w=[]
for i in range(d):
    val=(random.randint(L[i], G[i]))
    w.append(val)  

learning_rate= 0.2
count=0
while count <100:
    print(w)    
    sys.stdout.flush()
    fx=float(input())

    for i in range (0,d):
        f_x=float(input())
        w[i]= w[i]- learning_rate * f_x

    count=count-1
```
