```python
import numpy as np
import random
```

```python
def generate_data(n):
    x=np.sort(np.random.randn(n)*2)
    f=(np.cos(x)+2)/(np.cos(1.4*x)+2)
    noise=np.random.normal(0,1,n)*np.std(f)* 0.5
    y=f+noise
    return x,y,f
```

```python
x,y,f=generate_data(100)
print(x)
x=x.reshape(100,1)
print(x)
```

```python
print(y)
y=y.reshape(100,1)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y, marker='.', linewidths=1.5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(X_train)
print(X_train.shape)
```

```python
print(y_train)
```

```python
print(X_test.shape)
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
```

```python
clf = Ridge(alpha=0.001)
clf.fit(X_train, y_train)
plt.plot(X_test, clf.predict(X_test), linewidth=1.5,c = "red")
plt.scatter(x,y)
```

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
plt.plot(X_test, regr.predict(X_test),  linewidth=1.5,c = "green")
plt.scatter(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
def poly_f(x):
    poly_features = PolynomialFeatures(degree=10, include_bias=False)
    x=poly_features.fit_transform(x)
    return x
```

```python
alpha = 0.05
X_train_poly= poly_f(X_train)
n, m = X_train_poly.shape
I = np.identity(m)
w = np.dot(np.dot(np.linalg.inv(np.dot(X_train_poly.T, X_train_poly) + alpha * I), X_train_poly.T), y_train)
print(w)
```

```python
print(w.shape)
```

```python
X_test_poly=poly_f(X_test)
print(X_test_poly.shape)
```

```python
y_pred=np.dot(w.T,X_test_poly.T)
y_pred=y_pred.T
sort_zip=sorted(zip(X_test,y_pred))
a,b=zip(*sort_zip)
plt.plot(a,b,color='red')
plt.scatter(x,y)
```

```python
alpha = 0.01
X_train_poly= poly_f(X_train)
n, m = X_train_poly.shape
I = np.identity(m)
w = np.dot(np.dot(np.linalg.inv(np.dot(X_train_poly.T, X_train_poly) + alpha * I), X_train_poly.T), y_train)
X_test_poly=poly_f(X_test)
y_pred=np.dot(w.T,X_test_poly.T)
y_pred=y_pred.T
sort_zip=sorted(zip(X_test,y_pred))
a,b=zip(*sort_zip)
plt.plot(a,b,color='red')
plt.scatter(x,y)
```

```python
regr = LinearRegression()
X_train_poly_r= poly_f(X_train)
regr.fit(X_train_poly_r, y_train)
X_test_poly_r=poly_f(X_test)
y_pred_r=regr.predict(X_test_poly_r)
sort_zip=sorted(zip(X_test,y_pred_r))
a,b=zip(*sort_zip)
plt.plot(a,b,color='green')
plt.scatter(x,y)
```
