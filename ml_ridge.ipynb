```python
import numpy as np
import matplotlib.pyplot as plt
import array
import math

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

```python
x_r = np.sort(np.random.randn(100)*2)
print(x_r.shape)
f=(np.cos(x_r)+2)/(np.cos(1.4*x_r)+2)
#print(f)
noise = np.random.normal(0, 1, 50) * np.std(f) * 0.5;
y = f + noise;
#print(y)
#f = (cos(x) + 2) ./ (cos(1.4 .* x) + 2);
#print(x)
#x_r = x_r.reshape(500, 1)
#f = f.reshape(500, 1)
plt.scatter(x_r,f)
```

```python
x_r = x_r.reshape(500, 1)
#print(x_r.shape)
```

```python
x_ = np.random.normal(0, 1, 500)
x_.sort()
print(x_.shape)
y_=(np.cos(x_)+2)/(np.cos(1.4*(x_)+2))
#print(y_)
x_ = x_.reshape(500, 1)
#print(x_)
```

```python
x_n = np.concatenate((x_r, x_), axis=0)
print(x_n.shape)
#print(x_n)
y_n = np.concatenate((y, y_), axis=0)
y_n = y_n.reshape(1000, 1)
print(y_n.shape)
#print(y_n)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_n, y_n, test_size=0.2, random_state=0)
```

```python
clf = Ridge(alpha=0.005)
clf.fit(X_train, y_train)
plt.plot(X_test, clf.predict(X_test), linewidth=1.5)
plt.scatter(X_test,y_test)
regr = LinearRegression()
regr.fit(X_train, y_train)
plt.plot(X_test, regr.predict(X_test),  linewidth=1.5)
```
