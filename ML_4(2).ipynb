```python
import numpy as np
import random
```

```python
def generate_data(n):
    x=np.sort(np.random.randn(n)*2)
    f=(np.cos(x)+2)/(np.cos(1.4*x)+2)
    noise=np.random.normal(0,1,n)*np.std(f)* 0.5
    y=f+noise
    return x,y,f
```

Non Parametric RBF

```python
def non_param_RBF(r,X_train,X_test,y_train,y_test):
    y_t=[]
    for i in range(len(X_test)):
        x_i=X_test[i]
        up=0
        lower=0
        for j in range(len(X_train)):
            z=abs(x_i-X_train[j])/r
            phi=np.exp(-.5*(z**2))
            lower=lower+phi
            up=up+phi*y_train[j]
        g_x=up/lower
        y_t.extend(g_x)  
    #print(y_t)
    y_pred = np.array(y_t)
    y_pred=y_pred.reshape(len(X_test),1)
    return y_pred
```

Parametric RBF

```python
x,y,f=generate_data(100)
x=x.reshape(100,1)
y=y.reshape(100,1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
k = np.linspace(min(min(X_train),min(X_test)), max(max(X_train),max(X_test)),10)
print(k)
```

```python
def param_RBF(r,k,X_train, X_test, y_train, y_test):
    x_d=[]
    for i in range(len(X_train)):
        x_i=X_train[i]
        xn=[]
        for j in range(len(k)):
            xd=np.exp(-.5*(abs(x_i-k[j])/r)**2)
            xn.extend(xd)
        x_d.extend(xn)
        xdl=np.array(x_d)
    xdl=xdl.reshape(len(X_train),len(k))
    w=np.dot(np.dot(np.linalg.inv(np.dot(xdl.T,xdl)),xdl.T),y_train)
    
    x_t=[]
    for i in range(len(X_test)):
        x_i=X_test[i]
        xn=[]
        for j in range(len(k)):
            xd=np.exp(-.5*(abs(x_i-k[j])/r)**2)
            xn.extend(xd)
        x_t.extend(xn)
        xtl=np.array(x_t)
    xtl=xtl.reshape(len(X_test),len(k))

    y_pred = np.dot(w.T, xtl.T).T
    print(y_pred.shape)
    error = mean_squared_error(y_test, y_pred)
    return error
    
```

```python
error=param_RBF(0.2,k,X_train, X_test, y_train, y_test)
```

```python
print(error)
```
