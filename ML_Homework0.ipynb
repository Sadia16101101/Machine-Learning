```python
import numpy as np
A = np.array([[3, 4, 5],[6 ,7, 8]])
B = np.array([[1, 2, 3],[4, 5, 6]])
  
print(A)
print(B)

b= B.transpose()
print(b)

```

```python
mul= np.dot(A,b)
print(mul)
```

```python
C = np.array([[1, 0],[0, 2]])
c= np.linalg.inv(C)
print(c)

add = mul.__add__(c)
print(add)
```
